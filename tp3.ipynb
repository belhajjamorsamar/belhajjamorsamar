```python
import pandas as pd

l = [ [1,120] ,[2,155] ,[3,125]  ,[4,202] ,[5,180],[6,235],[7,240]
      ]
df = pd.DataFrame(l,columns=['temps','CA'])
df
```

```python
import matplotlib.pyplot as plt
plt.scatter (df['temps'],df['CA'])
```

```python
a=(df['CA'][6]-df['CA'][0])/(df['temps'][6]-df['temps'][0])
a
b=df['CA'][6]-(a*df['temps'][6])
b
```

```python
def calcul (p1,p2):
   a=(p2[1]-p1[1])/(p2[0]-p2[0])
   b=(p2[1])-a*p2[0]
   return a,b
calcul([df['CA'][0],df['temps'][0]],[df['temps'][6],df['CA'][6]]) 

```
